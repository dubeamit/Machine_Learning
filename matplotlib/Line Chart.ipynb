```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(0,10,10)
x
```

```python
y = np.sin(x)
y
```

```python
plt.plot(x,y)
```

```python
plt.plot(x,y)
plt.xlabel('time')
plt.ylabel('func of time')
plt.title('PLOT')
plt.show()
```
